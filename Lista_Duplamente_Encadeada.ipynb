```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
    self.prev = None

class ListaDuplamenteEncadeada:
  def __init__(self):
    self.head = None

  def append(self, data):
    new_node = Node(data)

    if self.head == None:
      self.head = new_node
      new_node.prev = None
      return

    current_node = self.head

    while current_node.next:
      current_node = current_node.next

    current_node.next = new_node
    new_node.prev = current_node
    new_node.next = None
    return


  def display(self):
    node = self.head

    if node is None:
      print("Lista Vazia")

    while node:
      print(node.data)
      node = node.next
    print("----------------")

    def length(self):
      if self.head == None:
          return 0
      current_node = self.head
      total = 0

      while current_node:
          total += 1
          current_node = current_node.next
      return total


    def insert_at_start(self, data):
      new_node = Node(data)

      if self.head == None:
        self.head = new_node
        new_node.prev = None
        return

      new_node.next = self.head
      self.head.prev = new_node
      self.head = new_node


    def remove_at_end(self):
      if self.head == None:
        print("Lista vazia")
        return

      current_node = self.head

      while current_node.next != None:
        current_node = current_node.next

      current_node.prev.next = None


    def remove_at_start(self):
       if self.head == None:
            print("The list has no element to delete")
            return
        if self.head.next == None:
            self.head = None
            return

        self.head = self.head.next
        self.head.prev = None

     def remove_element_by_value(self, value):
        if self.head == None:
            print("The list has no element to delete")
            return
        if self.head.next == None:
            if self.head.item == value:
                self.head = None
            else:
                print("Item not found")
            return

        current_node = self.head

        while current_node.next != None:
          if current_node.data == value:
            break
          current_node = current_node.next

        if current_node.next != None:
          current_node.prev.next = current_node.next
          current_node.next.prev = current_node.prev
        else:
            if current_node.data == value:
                current_node.prev.next = None

my_list = ListaDuplamenteEncadeada()
my_list.display()
# Add the elements
my_list.append(3)
my_list.append(2)
my_list.append(7)
my_list.append(1)

my_list.display()
```
