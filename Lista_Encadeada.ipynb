```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

class ListaEncadeada:
  def __init__(self):
    self.head = None

  def append(self, data):
    new_node = Node(data)

    if self.head == None:
      self.head = new_node
      return

    current_node = self.head

    while current_node.next:
      current_node = current_node.next

    current_node.next = new_node
    return

  def lenght(self):
    if self.head == None:
      return 0

    current_node = self.head
    total = 0

    while current_node:
      total += 1
      current_node = current_node.next

    return total

  def display(self):
    node = self.head

    if node is None:
      print("Lista Vazia")

    while node:
      print(node.data)
      node = node.next
    print("----------------")


  def reverse_linkedlist(self):
    prev_node = None
    current_node = self.head

    while current_node != None:
      next = current_node.next
      current_node.next = prev_node
      prev_node = current_node
      current_node = next
    self.head = prev_node

  def get(self, index):
    if index >= self.lenght() or index < 0:
      print("Index fora dos limites")

    current_node = self.head
    cont = 0

    while current_node != None:
      if cont == index:
        return current_node.data
      current_node = current_node.next
      cont += 1

  def search_item(self, item):
    current_node = self.head

    if current_node is None:
      print("Lista Vazia")

    while current_node != None:
      if current_node.data == item:
        print("O elemento {} existe na lista" .format(item))
        return
      current_node = current_node.next
    print("O elemento {} não está na lista" .format(item))

  def remove_at_start(self):
    if self.head == None:
      print("Lista Vazia")
      return

    self.head = self.head.next

  def remove_at_end(self):
    if self.head == None:
      print("Lista Vazia")
      return

    current_node = self.head

    while current_node != None:
      current_node = current_node.next
    current_node.next = None

  def insert_at_start(self, data):

    new_node = Node(data)
    new_node.next = self.head
    self.head = new_node

  def insert_at_end(self, data):

    new_node = Node(data)

    if self.head == None:
      self.head = new_node
      return

    current_node = self.head

    while current_node != None:
      current_node = current_node.next
    current_node.next = new_node


  def remove_by_value(self, value):
    current_node = self.head

    if current_node == None:
      return
    elif current_node.data == value:
      self.head = current_node.next
      current_node = None
      return

    while current_node != None:
      if current_node.data == value:
        break
      prev = current_node
      current_node = current_node.next

    if current_node == None:
      return

    prev.next = current_node.next
    current_node = None

  def insert_at_index(self, data, index):
    new_node = Node(data)

    if self.head == None:
      self.head = new_node
      return

    if index >= self.lenght() or index < 0:
      print("Posição fora dos limites")
      return

    current_node = self.head
    pos = 0

    while pos < index - 1 and current_node != None:
      current_node = current_node.next
      pos += 1

    new_node.next = current_node.next
    current_node.next = new_node


my_list = ListaEncadeada()
my_list.display()

my_list.append(3)
my_list.append(7)
my_list.append(2)
#9
my_list.append(1)
my_list.append(5)
my_list.append(4)
my_list.insert_at_index(9, 3)
my_list.display()
```
